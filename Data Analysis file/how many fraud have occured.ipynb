```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
from datasets import load_dataset
import numpy as np

df = pd.read_csv('D:\.MY PERSONAL DATA/programming Practice/Data Analytics/Real Dataset Practice/Credit Card Fraud/Code of the project/Results Gathered/in which hour frauds have occured.csv')
```

```python

df.head()
```

```python
# --- Convert to long (tidy) format ---
df_long = df.melt(id_vars='hour', 
                  value_vars=['total_transactions', 'fraud_transactions'],
                  var_name='Transaction Type', 
                  value_name='Count')

# --- Plot with Seaborn ---
plt.figure(figsize=(16, 6))
sns.set(style="whitegrid")

sns.lineplot(
    data=df_long,
    x='hour',
    y='Count',
    hue='Transaction Type',
    marker='o',
    markersize=6,
    linewidth=2
)
```

```python

plt.figure(figsize=(10, 5))
sns.lineplot(data=df,x='hour',y='total_transactions',markers='o')
plt.tight_layout()
```

```python
sns.barplot(data=df,x='hour',y='total_transactions')
plt.figure(figsize=(50, 15))
plt.tight_layout()
```

```python
fig, ax1 = plt.subplots(figsize=(16, 6))
sns.set(style="whitegrid")

# Ploted total transactions
sns.lineplot(data=df, x='hour', y='total_transactions', marker='o', ax=ax1, label='Total Transactions', color='red')
ax1.set_ylabel('Total Transactions', color='blue')

# Created a second y-axis for fraud
ax2 = ax1.twinx()
sns.lineplot(data=df, x='hour', y='fraud_transactions', marker='o', ax=ax2, label='Fraud Transactions', color='green')
ax2.set_ylabel('Fraud Transactions', color='orange')

# Title and layout
plt.title('Total vs Fraud Transactions by Hour (Separate Scales)')
plt.xticks(range(0, 48, 2))
fig.tight_layout()
plt.show()
```
